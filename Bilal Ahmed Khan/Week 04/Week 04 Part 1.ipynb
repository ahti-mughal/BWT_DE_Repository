# Question 01

```python
def fibonacci(num):
    if(num==1):
        return 1
    if(num==0):
        return 0
    
    return fibonacci(num-1)+fibonacci(num-2)
```

```python
print(fibonacci(10))
```

# Question 04

```python
def fine_calculator(speed):
    if(speed <= 70):
        return "No fine"
    elif(speed>70 and speed<=80):
        return "Less fine"
    elif (speed>80):
        return "Car Sieze"
```

```python
import random

speed=random.randint(0,125)

print("Your speed is",speed,"\b. Your fine will be ", fine_calculator(speed),"\b.")
```

# Question 05

```python
def duplicate_exist(array):
    duplicate=False
    for i in range(0,len(array)):
        if(array.count(array[i]) > 1):
            array.remove(array[i])
            duplicate=True
            break
    if(duplicate):       
        return True
    else:
        return False
```

```python
import random
array =[]

for i in range (0,10):
    array.append(random.randint(0,100))

print(array)
if(duplicate_exist(array)):
    print("\nYes duplicates exist in the given array.")
    print("After removing duplicate element we have\n")
    print(array)
else:
    print("No duplicates exist in the given array.")
```

# Question 07

```python
text = input("Enter the text ")

count_letter=0
count_number=0

for i in range (0,len(text)):
    if((ord(text[i])>=65 and ord(text[i])<=90) or (ord(text[i])>=97 and ord(text[i])<=122)):
        count_letter+=1
    elif((ord(text[i])>=48 and ord(text[i])<=57)):
        count_number+=1
        
print("The text you entered is",text)
print("Count of numbers in the given text is ", count_letter,"\b.")
print("Count of letters in the given text is ", count_number,"\b.")
```
